```python
import os
import csv
from pathlib import Path

base = Path("../dataset_squares")
rows = []
id_counter = 1


types = {
    "empty" : 0,

    "black_king" : 1,
    "black_queen" : 2,
    "black_rook" : 3,
    "black_bishop" : 4,
    "black_knight" : 5,
    "black_pawn" : 6,

    "white_king" : 7,
    "white_queen" : 8,
    "white_rook" : 9,
    "white_bishop" : 10,
    "white_knight" : 11,
    "white_pawn" : 12
}
for figure_type_dir in sorted(base.iterdir()):
    if not figure_type_dir.is_dir():
        continue
    name = figure_type_dir.name
    if name in types:
        figure_type = types[name]
    else:
        continue
    
    for img_path in sorted(figure_type_dir.glob("*.jpg")):
        rows.append({
            "id": f"P{id_counter:05d}",
            "image_path": str(img_path),
            "label": figure_type
        })
        id_counter += 1

# zapis CSV
with open("train.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=["id", "image_path", "label"])
    writer.writeheader()
    writer.writerows(rows)

print(f"Generowano {len(rows)} przykładow")
```
